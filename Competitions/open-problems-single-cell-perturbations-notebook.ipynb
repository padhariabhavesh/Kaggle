# Import Library

```python
import pandas as pd
```

# Step 1: Load the training data

```python
train_path = '/kaggle/input/open-problems-single-cell-perturbations/de_train.parquet'
df_train = pd.read_parquet(train_path)
```

```python
# Print the training data if necessary
df_train
```

# Step 2: Group by drug and calculate the mean

```python
df_simple = df_train.iloc[:, [1] + list(range(5, df_train.shape[1]))]
mean_df = df_simple.groupby('sm_name').mean().reset_index()
mean_df
```

# Step 3: Load ID map

```python
id_map_path = '/kaggle/input/open-problems-single-cell-perturbations/id_map.csv'
df_ids = pd.read_csv(id_map_path)
print(df_ids.shape)
display(df_ids)  # This is where you are trying to access df_ids
```

# Step 4: Load the sample template

```python
sample_submission_path = '/kaggle/input/open-problems-single-cell-perturbations/sample_submission.csv'
submit_df = pd.read_csv(sample_submission_path)
print(submit_df.shape)
display(submit_df)
```

# Step 5: Filter and prepare rows based on 'sm_name'

```python
filtered_rows = []

for variable_name in df_ids['sm_name']:
    matching_rows = mean_df[mean_df['sm_name'] == variable_name].copy()
    matching_rows['variable_name'] = variable_name
    filtered_rows.append(matching_rows)

# Concatenate all filtered rows into a single DataFrame
result_df = pd.concat(filtered_rows)
result_df = result_df.reset_index(drop=True)
result_df
```

# Step 6: Paste the results into the sample submission file

```python
for i, col in enumerate(submit_df.columns):
    if col == 'id':
        continue
    submit_df[col] = result_df[col]
    if (i % 1000) == 0:
        print(i, col)
```

# Step 7: Group by cell type and calculate the mean

```python
df_simple = df_train.iloc[:, [0] + list(range(5, df_train.shape[1]))]
cell_type_mean = df_simple.groupby('cell_type').mean().reset_index()
cell_type_mean
```

# Step 8: Filter and prepare rows based on 'cell_type'

```python
filtered_rows = []

for variable_name in df_ids['cell_type']:
    matching_rows = cell_type_mean[cell_type_mean['cell_type'] == variable_name].copy()
    matching_rows['variable_name'] = variable_name
    filtered_rows.append(matching_rows)

# Concatenate all filtered rows into a single DataFrame
result_df = pd.concat(filtered_rows)
result_df = result_df.reset_index(drop=True)
result_df
```

# Step 9: Load the sample template for cell type

```python
submit_df_celltype = pd.read_csv(sample_submission_path)
submit_df_celltype
```

# Step 10: Paste the results into the sample submission file for cell type

```python
for i, col in enumerate(submit_df_celltype.columns):
    if col == 'id':
        continue
    submit_df_celltype[col] = result_df[col]
    if (i % 1000) == 0:
        print(i, col)
```

# Step 11: Combine the submission with a high-scoring one

```python
fn = '/kaggle/input/open-problems-single-cell-perturbations/sample_submission.csv'
v18sub = pd.read_csv(fn)
print(v18sub.shape)


# Adjust the weights (0.45, 0.45, 0.1) as needed
average_df = v18sub * 0.45 + submit_df * 0.45 + submit_df_celltype * 0.1 

# Convert 'id' back to int32
average_df['id'] = average_df['id'].astype('int32')

# Save the final submission as a CSV file
average_df.to_csv('submission.csv', index=False)
average_df
```
